###  b) Remplaçons les valeurs aberrantes inferieure à 100g par la median, mode etc..

#### b1) par mediane

```python
for cols in colms:
    Q1,Q3,IQR=af.DataFrame_and_description._interqartile_(Data_lessthan100,cols)
    Data_lessthan100,lower_whisker,upper_whisker=af.Dealing_with_defect._Drop_outliers(Data_lessthan100, cols,Q1,Q3,IQR)
   
    median=Data_lessthan100[cols].median()
    dataMedian=af.Dealing_with_defect._replacing_outliers_with_median(Data_lessthan100,median,cols,upper_whisker,lower_whisker)
```

#### b2) par mode

```python
for cols in colms:
    Q1,Q3,IQR=af.DataFrame_and_description._interqartile_(Data_lessthan100,cols)
    Data_lessthan100,lower_whisker,upper_whisker=af.Dealing_with_defect._Drop_outliers(Data_lessthan100, cols,Q1,Q3,IQR)
    mode=Data_lessthan100[cols].mode()
    dataMode=af.Dealing_with_defect._replacing_outliers_with_median(Data_lessthan100,mode,cols,upper_whisker,lower_whisker)
   
```
