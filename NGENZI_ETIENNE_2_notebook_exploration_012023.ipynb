```python
kmeans.inertia_
```

```python
intertia = []
```

```python
k_list = range(1, 10)
list(k_list)
```

```python
for i in k_list :
  kmeans = af.KMeans(n_clusters=i)
  kmeans.fit(n) 
  intertia.append(kmeans.inertia_)
```

```python
intertia
```

```python
fig, ax = af.plt.subplots(1,1,figsize=(12,6))

ax.set_ylabel("intertia")
ax.set_xlabel("n_cluster")
ax = af.plt.plot(k_list, intertia)
```
